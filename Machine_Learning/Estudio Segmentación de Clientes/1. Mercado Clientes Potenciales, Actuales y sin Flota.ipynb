### Parque Automotriz 

```python
# Parque automotriz usado en el modelo

df_parque = pd.read_csv(URL+'ParqueAutomotriz2020_Modelo.csv')
```

```python
# Nos quedamos con los vehículos con año de fabricación sobre el 2000
df_parque = df_parque[df_parque['ano_fab'] >= 2000]
```

### Dimensión Tipo de Vehículo

```python
df_tipo = pd.read_excel(URL + 'Dim_TipoVehiculo.xlsx', sheet_name='Hoja1')
df_tipo.columns = ['Tipo', 'Tipo_Corto','Clasificacion']
```

```python
df_tipo.head()
```

```python
# Filtramos clasificación Liviano y Pesado

df_ptv = pd.merge(df_parque, df_tipo, on = 'Tipo', how = 'left')
df_ptv = df_ptv[(df_ptv['Clasificacion'] == 'LIVIANO') | (df_ptv['Clasificacion'] == 'PESADO')]
```

```python
df_ptv['ppu'].count()
```

### Flota por Empresa

```python
df_ptv.head()
```

```python
# Q de Flota

# Transporte de Carga
df_ptv.loc[(df_ptv['Tipo'] == 'CAMION') | (df_ptv['Tipo'] == 'CHASIS CABINADO'), 'Q Camion'] = 1
df_ptv.loc[df_ptv['Tipo'] == 'CAMION LIVIANO', 'Q Camion Liv'] = 1
df_ptv.loc[df_ptv['Tipo'] == 'REMOLQUE', 'Q Remolque'] = 1
df_ptv.loc[df_ptv['Tipo'] == 'SEMIREMOLQUE', 'Q Semi Remolque'] = 1
df_ptv.loc[df_ptv['Tipo'] == 'TRACTOCAMION', 'Q Tractocamion'] = 1

# Transporte de Pasajeros
df_ptv.loc[df_ptv['Tipo'] == 'BUS', 'Q Bus'] = 1
df_ptv.loc[df_ptv['Tipo'] == 'LOC. COLECTIVA', 'Q Loc Colectiva'] = 1
df_ptv.loc[(df_ptv['Tipo'] == 'MINIBUS') | (df_ptv['Tipo'] == 'TAXIBUS'), 'Q Mini Bus'] = 1
df_ptv.loc[df_ptv['Tipo'] == 'FURGON', 'Q Furgon'] = 1

# Particular
df_ptv.loc[(df_ptv['Tipo'] == 'AUTOMOVIL') | (df_ptv['Tipo'] == 'JEEP') | (df_ptv['Tipo'] == 'STATION WAGON') 
           | (df_ptv['Tipo'] == 'TODO TERRENO'), 'Q Automovil'] = 1

df_ptv.loc[df_ptv['Tipo'] == 'CAMIONETA', 'Q Camioneta'] = 1
```

```python
df_ptv.head()
```

```python
df_ptv[df_ptv['rut'] == 76236918]
```

```python
df_tipoflota = df_ptv.groupby(['RUT_Final']).agg({
    'Q Camion': 'sum', 
    'Q Camion Liv': 'sum',
    'Q Remolque': 'sum',
    'Q Semi Remolque': 'sum',
    'Q Tractocamion': 'sum',
    'Q Bus': 'sum',
    'Q Loc Colectiva': 'sum',
    'Q Mini Bus': 'sum',
    'Q Furgon': 'sum',
    'Q Automovil': 'sum',
    'Q Camioneta': 'sum'
}).reset_index()


df_tipoflota['Q Camion'] = df_tipoflota['Q Camion'].astype('int64') 
df_tipoflota['Q Camion Liv'] = df_tipoflota['Q Camion Liv'].astype('int64')  
df_tipoflota['Q Remolque'] = df_tipoflota['Q Remolque'].astype('int64') 
df_tipoflota['Q Semi Remolque'] = df_tipoflota['Q Semi Remolque'].astype('int64') 
df_tipoflota['Q Tractocamion'] = df_tipoflota['Q Tractocamion'].astype('int64') 
df_tipoflota['Q Bus'] = df_tipoflota['Q Bus'].astype('int64') 
df_tipoflota['Q Loc Colectiva'] = df_tipoflota['Q Loc Colectiva'].astype('int64') 
df_tipoflota['Q Mini Bus'] = df_tipoflota['Q Mini Bus'].astype('int64') 
df_tipoflota['Q Furgon'] = df_tipoflota['Q Furgon'].astype('int64') 
df_tipoflota['Q Automovil'] = df_tipoflota['Q Automovil'].astype('int64') 
df_tipoflota['Q Camioneta'] = df_tipoflota['Q Camioneta'].astype('int64') 

df_tipoflota['Q Flota Carga'] = df_tipoflota['Q Camion'] + df_tipoflota['Q Camion Liv'] + df_tipoflota['Q Remolque'] + df_tipoflota['Q Semi Remolque'] + df_tipoflota['Q Tractocamion'] 
df_tipoflota['Q Flota T de Pasajeros'] = df_tipoflota['Q Bus'] + df_tipoflota['Q Loc Colectiva'] + df_tipoflota['Q Mini Bus'] + df_tipoflota['Q Furgon']
df_tipoflota['Q Flota Particular'] = df_tipoflota['Q Automovil'] + df_tipoflota['Q Camioneta']

df_tipoflota['% Flota Carga'] = df_tipoflota['Q Flota Carga'] / (df_tipoflota['Q Flota Carga'] + df_tipoflota['Q Flota T de Pasajeros'] + df_tipoflota['Q Flota Particular'])
df_tipoflota['% Flota T de Pasajeros'] = df_tipoflota['Q Flota T de Pasajeros'] / (df_tipoflota['Q Flota Carga'] + df_tipoflota['Q Flota T de Pasajeros'] + df_tipoflota['Q Flota Particular'])
df_tipoflota['% Flota Particular'] = df_tipoflota['Q Flota Particular'] / (df_tipoflota['Q Flota Carga'] + df_tipoflota['Q Flota T de Pasajeros'] + df_tipoflota['Q Flota Particular'])

df_tipoflota = df_tipoflota.loc[df_tipoflota['RUT_Final'] != 0]
```

```python
# Consumos

df_flota = df_ptv.groupby(['RUT_Final','Clasificacion']).agg({
    'ppu': 'count', 
    'Consumo_Teorico_Lubricantes': 'sum',
    'Consumo_Teorico_Repuestos': 'sum',
    'Consumo_Teorico_Neumaticos': 'sum'}).reset_index()
```

```python
df_flota[df_flota['RUT_Final'] == 74687374]
```

```python
df_fc = pd.pivot_table(df_flota, 
                       values = ['ppu','Consumo_Teorico_Lubricantes','Consumo_Teorico_Repuestos','Consumo_Teorico_Neumaticos'],
                       index = ['RUT_Final'], 
                       columns = ['Clasificacion'], 
                       aggfunc = {'ppu': np.sum,
                                 'Consumo_Teorico_Lubricantes': np.sum,
                                 'Consumo_Teorico_Repuestos': np.sum,
                                 'Consumo_Teorico_Neumaticos': np.sum}, 
                       fill_value = 0).reset_index()
```

```python
df_fc.head()
```

```python
df_fc = df_fc.droplevel(level = 0, axis = 1)
```

```python
df_fc.columns = ['RUT_Final','CT_Lub_Liviano','CT_Lub_Pesado','CT_Neu_Liviano','CT_Neu_Pesado','CT_Rep_Liviano','CT_Rep_Pesado','Flota_Liviano','Flota_Pesado']
```

```python
# Borramos RUT_Final = 0

df_fc = df_fc[df_fc['RUT_Final'] != 0]
```

```python
df_fc.head(20)
```

```python
# Potencial total

df_fc['CT_Total'] = df_fc['CT_Lub_Liviano'] + df_fc['CT_Lub_Pesado'] + df_fc['CT_Neu_Liviano'] + df_fc['CT_Neu_Pesado'] + df_fc['CT_Rep_Liviano'] + df_fc['CT_Rep_Pesado']

df_fc['CT_Total_Liv'] = df_fc['CT_Lub_Liviano'] + df_fc['CT_Neu_Liviano'] + df_fc['CT_Rep_Liviano']

df_fc['CT_Total_Pes'] = df_fc['CT_Lub_Pesado'] + df_fc['CT_Neu_Pesado'] + df_fc['CT_Rep_Pesado']

# Flota Total

df_fc['Flota_Total'] = df_fc['Flota_Liviano'] + df_fc['Flota_Pesado'] 
```

```python
df_fcf = pd.merge(df_fc,df_tipoflota, on = 'RUT_Final', how = 'left')
```
