### Consolidado

```python
# Left Join

df_fcv = pd.merge(df_fc, df_v, on = 'RUT_Final', how = 'left').fillna(0)
df_fcv['Venta_36_meses'] =  df_fcv['Venta_36_meses'].astype(int)


# Left Join - 24 meses

df_fcv = pd.merge(df_fcv, df_v24, on = 'RUT_Final', how = 'left').fillna(0)
df_fcv['Venta_24_meses'] =  df_fcv['Venta_24_meses'].astype(int)


# Left Join - 12 meses

df_fcv = pd.merge(df_fcv, df_v12, on = 'RUT_Final', how = 'left').fillna(0)
df_fcv['Venta_12_meses'] =  df_fcv['Venta_12_meses'].astype(int)
```

```python
df_fcv.head()
```

```python
df_fcv[df_fcv['RUT_Final'] == 94528000]
```

```python
# Condiciones inactivo - fugado

df_fcv.loc[(df_fcv['Venta_12_meses'] == 0) & (df_fcv['Venta_24_meses'] > 0),'Estado Cliente'] = 'Inactivo'
df_fcv.loc[(df_fcv['Venta_12_meses'] == 0) & (df_fcv['Venta_24_meses'] == 0) & (df_fcv['Venta_36_meses'] > 0) ,'Estado Cliente'] = 'Fugado'
```
