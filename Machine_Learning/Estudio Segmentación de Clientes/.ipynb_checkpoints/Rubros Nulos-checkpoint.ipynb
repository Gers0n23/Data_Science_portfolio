### Parque

```python
URL_P = 'C:/Users/juan.vergara/OneDrive - Caren Repuestos Flotacentro/Documentos/Análisis/Parque Automotriz/Parque en Bruto para Modelo/'

df_parque = pd.read_csv(URL_P + 'ParqueAutomotriz2020_Modelo.csv')
```

```python
df_parque = df_parque[df_parque['ano_fab'] >= 2000]
df_parque = df_parque[df_parque['RUT_Final'] != 0]

df_pbase = df_parque[['ppu','Tipo','RUT_Final']]
df_pbase.head()
```

```python
df_dimtipo = pd.read_excel(URL_P + 'Dim_TipoVehiculo.xlsx', sheet_name = 'Hoja1')
df_dimtipo = df_dimtipo[['tipo','clasificacion']]
df_dimtipo.columns = ['Tipo', 'Clasificacion']
df_dimtipo
```

```python
# Left Join

df_pbase = pd.merge(df_pbase, df_dimtipo, on = 'Tipo', how = 'left')
df_pbase.head()
```

```python
df_pbase = df_pbase.groupby(['RUT_Final','Tipo','Clasificacion']).agg({'ppu': 'count'}).reset_index()
```

```python
df_pbase = df_pbase[df_pbase['Clasificacion'] != 'OTRO']
df_pbase.head()
```

```python
df_pbasepv = pd.pivot_table(df_pbase, 
                            values = ['ppu'], 
                            index = ['RUT_Final'], 
                            columns = ['Clasificacion','Tipo'], 
                            aggfunc = {'ppu': np.sum}, 
                            fill_value = 0).reset_index()
```

```python
df_pbasepv.head(10)
```

```python
df_pbasepv = df_pbasepv.droplevel(level = 0, axis = 1)
```

```python
df_pbasepv.columns = ['RUT_Final','Liv_Auto','Liv_Camioneta','Liv_Furgon','Liv_Jeep','Liv_Station','Liv_TodoTerreno',
                     'Pes_Bus','Pes_Camion','Pes_CamionLiv','Pes_Loc','Pes_Minibus','Pes_Remolque']
```

### Rubros

```python
df_pbasepv.loc[(df_pbasepv['Pes_Camion'] + df_pbasepv['Pes_CamionLiv'] + df_pbasepv['Pes_Remolque']) > 0 ,'Rubros'] = 'TRANSPORTE' 
df_pbasepv.loc[(df_pbasepv['Pes_Camion'] + df_pbasepv['Pes_CamionLiv'] + df_pbasepv['Pes_Remolque']) > 0 ,'Actividad'] = 'TRANSPORTE DE CARGA' 

df_pbasepv.loc[(df_pbasepv['Pes_Bus'] + df_pbasepv['Pes_Loc'] + df_pbasepv['Pes_Minibus']) > 0 ,'Rubros'] = 'TRANSPORTE'
df_pbasepv.loc[(df_pbasepv['Pes_Bus'] + df_pbasepv['Pes_Loc'] + df_pbasepv['Pes_Minibus']) > 0 ,'Actividad'] = 'TRANSPORTE DE PASAJEROS'
```

### Subgrupos

```python
df_pbasepv.loc[(df_pbasepv['Pes_Camion'] + df_pbasepv['Pes_CamionLiv'] + df_pbasepv['Pes_Remolque']) > 
               (df_pbasepv['Pes_Bus'] + df_pbasepv['Pes_Loc'] + df_pbasepv['Pes_Minibus']),'Sub Rubros'] = 'TRANSPORTE DE CARGA' 

df_pbasepv.loc[(df_pbasepv['Pes_Camion'] + df_pbasepv['Pes_CamionLiv'] + df_pbasepv['Pes_Remolque']) < 
               (df_pbasepv['Pes_Bus'] + df_pbasepv['Pes_Loc'] + df_pbasepv['Pes_Minibus']),'Sub Rubros'] = 'TRANSPORTE DE PASAJEROS' 

df_pbasepv.loc[(df_pbasepv['Rubros'] != 'TRANSPORTE')  & ((df_pbasepv['Liv_Auto'] + df_pbasepv['Liv_Camioneta'] + df_pbasepv['Liv_Furgon'] +
                df_pbasepv['Liv_Jeep'] + df_pbasepv['Liv_Station'] + df_pbasepv['Liv_TodoTerreno']) <= 4) ,'Sub Rubros'] = 'RETAIL' 

df_pbasepv.loc[(df_pbasepv['Rubros'] != 'TRANSPORTE') & ((df_pbasepv['Liv_Auto'] + df_pbasepv['Liv_Camioneta'] + df_pbasepv['Liv_Furgon'] +
                df_pbasepv['Liv_Jeep'] + df_pbasepv['Liv_Station'] + df_pbasepv['Liv_TodoTerreno']) > 4) ,'Sub Rubros'] = 'OTROS TIPOS DE COMERCIO'
```
