```python
import pandas as pd
```

```python
URL = 'C:/Users/juan.vergara/OneDrive - Caren Repuestos Flotacentro/Documentos/Análisis/Parque Automotriz/Parque en Bruto para Modelo/'
```

```python
#df_ventas = pd.read_csv(URL + 'VentasEne2016Nov2021.csv', sep = ';')
df_ventas = pd.read_csv(URL + 'VentasEne2016Dic2021.csv', sep = ',')

df_ventas = df_ventas[df_ventas['Mes'].notnull()]

df_ventas['Año'] = df_ventas['Año'].astype('int64')
df_ventas['Mes'] = df_ventas['Mes'].astype('int64')

df_ventas['Periodo'] = df_ventas['Año'].astype(str) + '-' + df_ventas['Mes'].astype(str)
```

```python
df_ventas.loc[(df_ventas['Periodo'] == '2021-12') | 
             (df_ventas['Periodo'] == '2021-11') |
             (df_ventas['Periodo'] == '2021-10') |
             (df_ventas['Periodo'] == '2021-9') |
             (df_ventas['Periodo'] == '2021-8') |
             (df_ventas['Periodo'] == '2021-7') |
             (df_ventas['Periodo'] == '2021-6') |
             (df_ventas['Periodo'] == '2021-5') |
             (df_ventas['Periodo'] == '2021-4') |
             (df_ventas['Periodo'] == '2021-3') |
             (df_ventas['Periodo'] == '2021-2') |
             (df_ventas['Periodo'] == '2021-1'), 'TipoPeriodo'] = '12 meses moviles'

df_ventas.loc[(df_ventas['Periodo'] == '2020-12') | 
             (df_ventas['Periodo'] == '2020-11') |
             (df_ventas['Periodo'] == '2020-10') |
             (df_ventas['Periodo'] == '2020-9') |
             (df_ventas['Periodo'] == '2020-8') |
             (df_ventas['Periodo'] == '2020-7') |
             (df_ventas['Periodo'] == '2020-6') |
             (df_ventas['Periodo'] == '2020-5') |
             (df_ventas['Periodo'] == '2020-4') |
             (df_ventas['Periodo'] == '2020-3') |
             (df_ventas['Periodo'] == '2020-2') |
             (df_ventas['Periodo'] == '2020-1'), 'TipoPeriodo'] = '13 a 24 meses moviles'

df_ventas.loc[(df_ventas['Periodo'] == '2019-12') | 
             (df_ventas['Periodo'] == '2019-11') |
             (df_ventas['Periodo'] == '2019-10') |
             (df_ventas['Periodo'] == '2019-9') |
             (df_ventas['Periodo'] == '2019-8') |
             (df_ventas['Periodo'] == '2019-7') |
             (df_ventas['Periodo'] == '2019-6') |
             (df_ventas['Periodo'] == '2019-5') |
             (df_ventas['Periodo'] == '2019-4') |
             (df_ventas['Periodo'] == '2019-3') |
             (df_ventas['Periodo'] == '2019-2') |
             (df_ventas['Periodo'] == '2019-1'), 'TipoPeriodo'] = '25 a 36 meses moviles'

df_ventas = df_ventas[df_ventas['TipoPeriodo'].notnull()]
```

```python
df_ventas['RUT'] = df_ventas['RUT'].str.split('-').str[0]
df_ventas['RUT'] = df_ventas['RUT'].astype('int64')
```

```python
df_ventas.head()
```

```python
# Grupos

df_gr = pd.read_excel(URL + 'VentasNov2020Oct2021.xlsx', sheet_name='RUT')
df_gr = df_gr[['Rut','Rut_Grupo']]
df_gr.columns = ['RUT','RUT_Final']


df_gr.head()
```

```python
df_ventasgr = pd.merge(df_ventas, df_gr, on = 'RUT', how = 'left')

df_ventasgr.loc[df_ventasgr['RUT_Final'].isnull(), 'RUT_Final'] = df_ventasgr['RUT']
df_ventasgr['RUT_Final'] = df_ventasgr['RUT_Final'].astype('int64')

df_ventasgr = df_ventasgr.drop('RUT',1)

df_ventasgr.head()
```

```python
# Clasificacion

df_cl = pd.read_excel(URL + 'VentasNov2020Oct2021.xlsx', sheet_name='CLASIFICACION')
df_cl.head()
```

```python
# Left Join Final

df_ventasgrcl = pd.merge(df_ventasgr, df_cl, on = 'LineaN3', how = 'left')
df_ventasgrcl['clasificacion'] = df_ventasgrcl['clasificacion'].fillna('PESADO') 
df_ventasgrcl.head()
```

```python
df_ventasgrcl.to_csv(URL + 'VentasEne2016Dic2021_Modelo.csv', index = False)
```
