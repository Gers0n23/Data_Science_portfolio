```python
# Carga

df_2019 = pd.read_table(URL + '2019.txt', sep = ';').drop(0)
df_2020 = pd.read_table(URL + '2020.txt', sep = ';').drop(0)
df_2021 = pd.read_table(URL + '2021.txt', sep = ';').drop(0)

# Último año móvil
#df_2020 = df_2020[df_2020.Mes >= 10]

frames = [df_2021, df_2020, df_2019]


df = pd.concat(frames)


# Limpieza
df = df[df['Fecha'] != '-']
df = df[df['Linea'] != '-']
```

### RUT Grupo

```python
# Carga de Rut grupo

df_rut = pd.read_excel(URL+'RUT_Grupos.xlsx')
df_rut.head()
```

```python
df_rut.dtypes
```

```python
df_rut['Rut'] = df_rut['Rut'].astype('object')
df_rut['Rut_Grupo'] = df_rut['Rut_Grupo'].astype(object)
df_rut.dtypes
```

### DF Ventas

```python
df.columns
```

```python
df['Linea'].drop_duplicates()
```

```python
df_lineas = df[['RUT','Linea','N° Folio']]
df_lineas.head()
```

```python
# RUT

df_lineas['Rut'] = df_lineas['RUT'].str.split('-').str[0]
df_lineas['Rut'] = df_lineas['Rut'].astype('int64')

df_lineas.head()
```

```python
df_lineas.count()
```

```python
df_lineas.dtypes
```

```python
# Left Join
df_lineasrut = pd.merge(df_lineas, df_rut, on = 'Rut', how = 'left')

# If
df_lineasrut.loc[df_lineasrut['Rut_Grupo'].isnull(), 'RUT_v'] = df_lineasrut['Rut']
df_lineasrut.loc[df_lineasrut['Rut_Grupo'].notnull(), 'RUT_v'] = df_lineasrut['Rut_Grupo']

df_lineasrut.head()
```

```python
df_lineasrut = df_lineasrut.drop('RUT',1)
df_lineasrut = df_lineasrut.drop('Rut',1)
df_lineasrut = df_lineasrut.drop('Rut_Grupo',1)
df_lineasrut = df_lineasrut.drop('grupo',1)
```

```python
df_lineasrut['RUT'] = df_lineasrut['RUT_v']
df_lineasrut = df_lineasrut.drop('RUT_v',1)
```

```python
df_lineasrut.head()
```

```python
df_lineasrut['N° Folio'] = df_lineasrut['N° Folio'].astype('object')
```

```python
# Tabla Pivote

df_pt = pd.pivot_table(df_lineasrut, index = ['RUT','Linea'], values = 'N° Folio', aggfunc = 'count').reset_index()
```

```python
df_pt['Transac Prod'] = df_pt['N° Folio']
df_pt = df_pt.drop('N° Folio',1)
df_pt
```

### Transacciones Totales

```python
df_tt = pd.pivot_table(df_lineasrut, index = ['RUT'], values = 'N° Folio', aggfunc = 'count').reset_index()
```

```python
df_tt
```

### Join

```python
df_transaclinea = df_pt.join(df_tt.set_index('RUT'), on = 'RUT')
```

```python
df_transaclinea['Penetracion Linea'] = df_transaclinea['Transac Prod']/df_transaclinea['N° Folio']
df_transaclinea['Penetracion Linea'] = np.round(df_transaclinea['Penetracion Linea'],2)
```

```python
df_penetracion = df_transaclinea[['RUT','Linea','Penetracion Linea']]
```

```python
df_penetracion.head()
```

```python
# Pivotear tabla

df_penetracion_pt = pd.pivot_table(df_penetracion, values = 'Penetracion Linea', index=['RUT'], columns = 'Linea').reset_index().fillna(0)
df_penetracion_pt.head(50)
```

```python
df_penetracion_pt.count()
```

```python
df_penetracion_pt.to_csv('Penetracion_Transaccion.csv', index = False)
```
