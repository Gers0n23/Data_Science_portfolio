### RFM

```python
# Ubicación archivo
URL = 'C:/Users/juan.vergara/OneDrive - Caren Repuestos Flotacentro/Documentos/Análisis/Segmentación Clientes/Fase 3 Recopilación de Datos/Extracción de Datos/'
```

```python
# Carga de archivo eliminando la fila 0 (viene nulla)
df2020 = pd.read_table(URL+'2020.txt', sep = ';').drop(0)
df2021 = pd.read_table(URL+'2021.txt', sep = ';').drop(0)

df2020 = df2020[df2020.Mes >= 10]

frames = [df2020, df2021]

df_1 = pd.concat(frames)

# Limpiando los datos '-' del campo fecha
df_1 = df_1[df_1.Fecha != '-']
```

```python
df2021.head()
```

```python
df_1.columns
```

```python
df_1.head(5)
```

```python
# Nos quedamos con los campos relevantes
df_2 = df_1[['Fecha','RUT','Cliente','$ Venta','$ Margen','N° Folio']]
```

```python
df_2.dtypes
```

```python
# Formato de fecha
df_2.Fecha = pd.to_datetime(df_2.Fecha, format = '%d-%m-%Y')

# Limpieza del campo $ Venta
df_2['$ Venta'] = df_2['$ Venta'].str.replace('$', '')
df_2['$ Venta'] = df_2['$ Venta'].str.replace('.', '')
df_2['$ Venta'] = pd.to_numeric(df_2['$ Venta'])

# Limpieza del campo $ Margen
df_2['$ Margen'] = df_2['$ Margen'].str.replace('$', '')
df_2['$ Margen'] = df_2['$ Margen'].str.replace('.', '')
df_2['$ Margen'] = pd.to_numeric(df_2['$ Margen'])

df_2.head()
```

```python
# Separamos el campo RUT
df_2['Rut'] = df_2['RUT'].str.split('-').str[0]
df_2['Rut'] = df_2['Rut'].astype('int64')

df_2.head()
```

### RUT de Grupos

```python
# RUT GRUPOS

df_rut = pd.read_excel(URL+'RUT_Grupos.xlsx')
df_rut.head()
```

```python
df_rut.dtypes
```

```python
# Left Join, para traer el RUT del grupo

df_3 = pd.merge(df_2, df_rut, on ='Rut', how = 'left')

# If para rellenar la columna de RUT de grupo

df_3.loc[df_3['Rut_Grupo'].isnull(), 'RUT_v'] = df_3['Rut']
df_3.loc[df_3['Rut_Grupo'].notnull(), 'RUT_v'] = df_3['Rut_Grupo']

df_3.head()
```

```python
# Limpiamos el df

df_3 = df_3.drop('RUT',1)
df_3 = df_3.drop('Rut',1)
df_3 = df_3.drop('grupo',1)
df_3 = df_3.drop('Rut_Grupo',1)
```

```python
df_3['RUT'] = df_3['RUT_v']
df_3 = df_3.drop('RUT_v',1)
```

```python
df_3.head()
```

```python
# Tabla dinámica para resumir la última compra y el monto total de venta por cliente

df_4 = pd.pivot_table(df_3, index=['RUT'], values=['Fecha','$ Venta','$ Margen','N° Folio'],
                        aggfunc={'Fecha': np.max, '$ Venta':np.sum, '$ Margen':np.sum}).reset_index()
```

```python
# Incluimos el día de hoy para obtener recencia

df_4['Today'] = '2021-10-01'
#df_4['Today'] = pd.Timestamp.today().strftime('%d-%m-%Y')
df_4['Today'] = pd.to_datetime(df_4.Today)
```

```python
df_4.count()
```

### Recencia

```python
df_4['Recencia'] = df_4.Today - df_4.Fecha
```

```python
df_4.head()
```

```python
df_4['Recencia'] = df_4['Recencia'].astype('string')
df_4['Recencia'] = df_4['Recencia'].str.split(' ').str[0]
```

```python
df_4['Recencia'] = df_4['Recencia'].astype('int')
```

```python
df_4['Recencia'] = df_4['Recencia'].astype('int')
```

### Frecuencia (Nro Folios)

```python
# Nos quedamos sólo con las varables relevantes
df_5 = df_3[['RUT','N° Folio']]
df_5.head()
```

```python
# Borramos las filas duplicadas
df_5.drop_duplicates(subset=['N° Folio'])
```

```python
# Tabla dinámica contando el número de folios por cliente
df_5 = pd.pivot_table(df_5, index=['RUT'], values=['N° Folio'], aggfunc=np.count_nonzero).reset_index()
df_5.head()
```

```python
# Renombramos la columna
df_5['Frecuencia'] = df_5['N° Folio']
df_5 = df_5.drop('N° Folio',1)
df_5.head()
```

```python
df_5.count()
```

```python
# JOIN
```

```python
df_final = df_5.join(df_4.set_index('RUT'), on = 'RUT')
```

```python
df_final['Ticket_Promedio'] = df_final['$ Venta']/df_final['Frecuencia']
```

```python
df_final['Ticket_Promedio'] = np.round(df_final['Ticket_Promedio'],0)
```
