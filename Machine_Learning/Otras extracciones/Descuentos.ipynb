### ETL para Exploración

```python
import pandas as pd
import numpy as np
```

### Descuentos

```python
# Dirección archivo
URL = 'C:/Users/juan.vergara/OneDrive - Caren Repuestos Flotacentro/Documentos/Análisis/Segmentación Clientes/Fase 3 Recopilación de Datos/Extracción de Datos/'
```

```python
# Carga de archivo eliminando la fila 0 (viene nulla)
df_2021 = pd.read_table(URL+'2021.txt', sep = ';', ).drop(0)
df_2020 = pd.read_table(URL+'2020.txt', sep = ';', ).drop(0)
df_2019 = pd.read_table(URL+'2019.txt', sep = ';', ).drop(0)

frames = [df_2021, df_2020, df_2019]

df_1 = pd.concat(frames)

# Limpiando los datos '-' del campo fecha
df_1 = df_1[df_1.Fecha != '-']
```

```python
df_1.columns
```

```python
df_1.head(5)
```

```python
# Nos quedamos con los campos relevantes
df_2 = df_1[['RUT','$ Venta','$ Descuento']]
```

```python
df_2.dtypes
```

```python
# Limpieza del campo $ Venta
df_2['$ Venta'] = df_2['$ Venta'].str.replace('$','')
df_2['$ Venta'] = df_2['$ Venta'].str.replace('.','')
df_2['$ Venta'] = pd.to_numeric(df_2['$ Venta'])

# Limpieza del campo $ Descuento
df_2['$ Descuento'] = df_2['$ Descuento'].str.replace('$','')
df_2['$ Descuento'] = df_2['$ Descuento'].str.replace('.','')
df_2['$ Descuento'] = pd.to_numeric(df_2['$ Descuento'])

df_2.head()
```

```python
# Separamos el campo RUT
df_2['Rut'] = df_2['RUT'].str.split('-').str[0]
df_2['Rut'] = df_2['Rut'].astype('int64')

df_2.head()
```

### RUT de Grupos

```python
# RUT GRUPOS

df_rut = pd.read_excel(URL+'RUT_Grupos.xlsx')
df_rut.head()
```

```python
df_rut.dtypes
```

```python
# Left Join, para traer el RUT del grupo

df_3 = pd.merge(df_2, df_rut, on ='Rut', how = 'left')

# If para rellenar la columna de RUT de grupo

df_3.loc[df_3['Rut_Grupo'].isnull(), 'RUT_v'] = df_3['Rut']
df_3.loc[df_3['Rut_Grupo'].notnull(), 'RUT_v'] = df_3['Rut_Grupo']

df_3.head()
```

```python
# Limpiamos el df

df_3 = df_3.drop('RUT',1)
df_3 = df_3.drop('Rut',1)
df_3 = df_3.drop('grupo',1)
df_3 = df_3.drop('Rut_Grupo',1)
```

```python
df_3['RUT'] = df_3['RUT_v']
df_3 = df_3.drop('RUT_v',1)
```

```python
df_3.head()
```

```python
# Tabla dinámica para resumir la última compra y el monto total de venta por cliente

df_4 = pd.pivot_table(df_3, index=['RUT'], values=['$ Venta','$ Descuento'],
                        aggfunc={'$ Venta':np.sum, '$ Descuento':np.sum}).reset_index()
```

```python
df_4.head()
```

```python
df_4.dtypes
```

```python
df_4['% Descuento'] = df_4['$ Descuento'] / (df_4['$ Venta'] + df_4['$ Descuento'])
df_4['% Descuento'] = np.round(df_4['% Descuento'],2)
df_4.head()
```

```python
df_4.to_csv('% Descuento.csv', index = False)
```
