### ETL para Exploración

```python
import pandas as pd
import numpy as np
```

```python
# Dirección archivo
URL = 'C:/Users/juan.vergara/OneDrive - Caren Repuestos Flotacentro/Documentos/Análisis/Segmentación Clientes/Fase 3 Recopilación de Datos/Extracción de Datos/'
```

```python
# carga de datos

df2021 = pd.read_table(URL+'2021.txt', sep = ';').drop(0)
df2020 = pd.read_table(URL+'2020.txt', sep = ';').drop(0)
df2019 = pd.read_table(URL+'2019.txt', sep = ';').drop(0)

frames = [df2019, df2020, df2021]
df = pd.concat(frames)

# Limpieza
df = df[df['Fecha'] != '-']
df = df[df['Linea'] != '-']
```

```python
# Carga de Rut Grupo

df_rut = pd.read_excel(URL+'RUT_Grupos.xlsx')
df_rut.head()
```

```python
df_rut.dtypes
```

```python
df_rut['Rut'] = df_rut['Rut'].astype(object)
df_rut['Rut_Grupo'] = df_rut['Rut_Grupo'].astype(object)
df_rut.dtypes
```

```python
# Original df

df.columns
```

```python
df['Linea'].drop_duplicates()
```

```python
df_lineas = df[['RUT','Linea','$ Venta']]
```

```python
# RUT

df_lineas['Rut'] = df_lineas['RUT'].str.split('-').str[0]
df_lineas['Rut'] = df_lineas['Rut'].astype('int64')

df_lineas.head()
```

```python
df_lineas.count()
```

```python
df_lineas.dtypes
```

```python
# Left Join

df_lineasrut = pd.merge(df_lineas, df_rut, on ='Rut', how = 'left')

# If

df_lineasrut.loc[df_lineasrut['Rut_Grupo'].isnull(), 'RUT_v'] = df_lineasrut['Rut']
df_lineasrut.loc[df_lineasrut['Rut_Grupo'].notnull(), 'RUT_v'] = df_lineasrut['Rut_Grupo']

df_lineasrut.head()
```

```python
df_lineasrut = df_lineasrut.drop('RUT',1)
df_lineasrut = df_lineasrut.drop('Rut',1)
df_lineasrut = df_lineasrut.drop('Rut_Grupo',1)
df_lineasrut = df_lineasrut.drop('grupo',1)

df_lineasrut.head()
```

```python
df_lineasrut['RUT'] = df_lineasrut['RUT_v']
df_lineasrut = df_lineasrut.drop('RUT_v',1)
```

```python
df_lineasrut.head()
```

```python
# Limpieza del campo $ Venta
df_lineasrut['$ Venta'] = df_lineasrut['$ Venta'].str.replace('$', '')
df_lineasrut['$ Venta'] = df_lineasrut['$ Venta'].str.replace('.', '')
df_lineasrut['$ Venta'] = pd.to_numeric(df_lineasrut['$ Venta'])
```

```python
df_lineasrut['$ Venta'] = pd.to_numeric(df_lineasrut['$ Venta'])
df_lineasrut.head()
```

```python
df_pt = pd.pivot_table(df_lineasrut, index = ['RUT','Linea'], values = '$ Venta', aggfunc=np.sum).reset_index()
```

```python
df_pt['Venta Linea'] = df_pt['$ Venta']
df_pt = df_pt.drop('$ Venta',1)
df_pt
```

### Venta Total

```python
df_vt = pd.pivot_table(df_lineasrut, index = ['RUT'], values = '$ Venta', aggfunc=np.sum).reset_index()
```

```python
df_vt
```

### Join

```python
df_ventalinea = df_pt.join(df_vt.set_index('RUT'), on = 'RUT')
```

```python
df_ventalinea['Penetracion Linea'] = df_ventalinea['Venta Linea']/df_ventalinea['$ Venta']
df_ventalinea['Penetracion Linea'] = np.round(df_ventalinea['Penetracion Linea'],2)
```

```python
df_ventalinea
```

```python
df_penetracion = df_ventalinea[['RUT','Linea','Penetracion Linea']]
```

```python
df_penetracion.head()
```

```python
# Pivotear tabla

df_penetracion_pt = pd.pivot_table(df_penetracion, values = 'Penetracion Linea', index=['RUT'], columns = 'Linea').reset_index().fillna(0)
df_penetracion_pt.head(50)
```

```python
df_penetracion_pt.count()
```

```python
df_penetracion_pt.to_csv('Penetracion.csv', index = False)
```
