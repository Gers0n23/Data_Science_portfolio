```python
import pandas as pd
import numpy as np
```

### RFM

```python
# Dirección archivo
URL = 'C:/Users/juan.vergara/OneDrive - Caren Repuestos Flotacentro/Documentos/Análisis/Segmentación Clientes/Fase 3 Recopilación de Datos/Extracción de Datos/'
```

```python
# Carga de archivo eliminando la fila 0 (viene nulla)
df = pd.read_table(URL+'2021.txt', sep = ';', ).drop(0)

# Limpiando los datos '-' del campo fecha
df = df[df.Fecha != '-']
```

```python
df.columns
```

```python
df.head(5)
```

```python
# Nos quedamos con los campos relevantes
df = df[['Fecha','RUT','Cliente','$ Venta','$ Margen','N° Folio']]
```

```python
df.dtypes
```

```python
# Formato de fecha
df.Fecha = pd.to_datetime(df.Fecha, format = '%d-%m-%Y')

# Limpieza del campo $ Venta
df['$ Venta'] = df['$ Venta'].str.replace('$', '')
df['$ Venta'] = df['$ Venta'].str.replace('.', '')
df['$ Venta'] = pd.to_numeric(df['$ Venta'])

# Limpieza del campo $ Margen
df['$ Margen'] = df['$ Margen'].str.replace('$', '')
df['$ Margen'] = df['$ Margen'].str.replace('.', '')
df['$ Margen'] = pd.to_numeric(df['$ Margen'])

df.head()
```

```python
df['Rut'] = df['RUT'].str.split('-').str[0]
df['Rut'] = df['Rut'].astype('int64')

df.head()
```

```python
# RUT GRUPOS

df_rut = pd.read_excel(URL+'RUT_Grupos.xlsx')
df_rut.head()
```

```python
df_rut['Rut_Grupo'] = df_rut['Rut_Grupo'].astype(object)
df_rut.dtypes
```

```python
# Left Join

df_fuc = pd.merge(df, df_rut, on ='Rut', how = 'left')

# If

df_fuc.loc[df_fuc['Rut_Grupo'].isnull(), 'RUT_v'] = df_fuc['Rut']
df_fuc.loc[df_fuc['Rut_Grupo'].notnull(), 'RUT_v'] = df_fuc['Rut_Grupo']

df_fuc.head()
```

```python
df_fuc = df_fuc.drop('RUT',1)
df_fuc = df_fuc.drop('Rut',1)
df_fuc = df_fuc.drop('grupo',1)
df_fuc = df_fuc.drop('Rut_Grupo',1)
```

```python
df_fuc['RUT'] = df_fuc['RUT_v']
df_fuc = df_fuc.drop('RUT_v',1)
```

```python
df_fuc.head()
```

```python
# df_fuc = pd.pivot_table(df, index=['NroCliente','Cliente'], values=['Fecha'], aggfunc=np.max)

# Tabla dinámica para resumir la última compra y el monto total de venta por cliente
df_fuc2 = pd.pivot_table(df_fuc, index=['RUT','Cliente'], values=['Fecha','$ Venta','$ Margen','N° Folio'],
                        aggfunc={'Fecha': np.max, '$ Venta':np.sum, '$ Margen':np.sum}).reset_index()
```

```python
# Incluimos el día de hoy para obtener recencia
df_fuc2['Today'] = pd.Timestamp.today().strftime('%d-%m-%Y')
df_fuc2['Today'] = pd.to_datetime(df_fuc2.Today)
```

```python
df_fuc2.head()
```

```python
df_fuc2.dtypes
```

```python
df_fuc2['Recencia'] = df_fuc2.Today - df_fuc2.Fecha
```

```python
df_fuc2.head()
```

#### Frecuencia (Nro Folios)

```python
# Nos quedamos sólo con las varables relevantes
df_freq = df_fuc[['RUT','N° Folio']]
df_freq.head()
```

```python
# Borramos las filas duplicadas
df_freq.drop_duplicates(subset=['N° Folio'])
```

```python
# Tabla dinámica contando el número de folios por cliente
df_freq = pd.pivot_table(df_fuc, index=['RUT'], values=['N° Folio'], aggfunc=np.count_nonzero).reset_index()
```

```python
df_freq['Frecuencia'] = df_freq['N° Folio']
df_freq = df_freq.drop('N° Folio',1)
df_freq.head()
```

```python
# JOIN
```

```python
df_final = df_freq.join(df_fuc2.set_index('RUT'), on = 'RUT')
```

```python
df_final = df_final.drop('Cliente',1)
df_final = df_final.drop('Fecha',1)
df_final = df_final.drop('Today',1)
```
