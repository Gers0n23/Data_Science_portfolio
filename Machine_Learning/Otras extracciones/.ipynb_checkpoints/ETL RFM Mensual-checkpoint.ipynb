### ETL RFM (Frecuencia en Meses)

```python
import pandas as pd
import numpy as np
```

```python
# Dirección archivos

URL = 'C:/Users/juan.vergara/OneDrive - Caren Repuestos Flotacentro/Documentos/Análisis/Segmentación Clientes/Fase 3 Recopilación de Datos/Extracción de Datos/'
```

```python
# Carga de archivo eliminando la fila 0 (viene nula)

df2021 = pd.read_table(URL+'2021.txt', sep = ';').drop(0)
df2020 = pd.read_table(URL+'2020.txt', sep = ';').drop(0)
df2019 = pd.read_table(URL+'2019.txt', sep = ';').drop(0)

frames = [df2021, df2020, df2019]
df_1 = pd.concat(frames)

# Limpiando los datos '-' del campo fecha
df_1 = df_1[df_1['Fecha'] != '-']
```

```python
df_1.head()
```

```python
df_1.columns
```

```python
# Nos quedamos con los campos relevantes
df_2 = df_1[['Año','Mes','Fecha', 'RUT', 'Cliente', '$ Venta','$ Margen','N° Folio']]
```

```python
df_2.dtypes
```

```python
# Enteros
df_2['Año'] = df_2['Año'].astype('int')
df_2['Mes'] = df_2['Mes'].astype('int')

# Formato Fecha

df_2['Fecha'] = pd.to_datetime(df_2['Fecha'], format = '%d-%m-%Y')

# Formato campo $Venta
df_2['$ Venta'] = df_2['$ Venta'].str.replace('$','')
df_2['$ Venta'] = df_2['$ Venta'].str.replace('.','')
df_2['$ Venta'] = pd.to_numeric(df_2['$ Venta'])

# Formato campo $Margen
df_2['$ Margen'] = df_2['$ Margen'].str.replace('$','')
df_2['$ Margen'] = df_2['$ Margen'].str.replace('.','')
df_2['$ Margen'] = pd.to_numeric(df_2['$ Margen'])

df_2.head(5)
```

```python
df_2['Periodo'] = df_2['Año'].astype('str')+'-'+df_2['Mes'].astype('str')
```

```python
df_2.head(5)
```

```python
# Separamos el campo RUT
df_2['Rut'] = df_2['RUT'].str.split('-').str[0]
df_2['Rut'] = df_2['Rut'].astype('int')

df_2.head(5)
```

### RUT de Grupos

```python
# RUT Grupos

df_rut = pd.read_excel(URL+'RUT_Grupos.xlsx')
df_rut.head()
```

```python
df_rut.dtypes
```

```python
# Left join, para traer el rut del grupo

df_3 = pd.merge(df_2, df_rut, on = 'Rut', how = 'left')

# If para rellenar la columna de RUT de grupo

df_3.loc[df_3['Rut_Grupo'].isnull(), 'RUT_v'] = df_3['Rut']
df_3.loc[df_3['Rut_Grupo'].notnull(), 'RUT_v'] = df_3['Rut_Grupo']


df_3.head()
```

```python
# Limpiamos el df

df_3 = df_3.drop('RUT',1)
df_3 = df_3.drop('Rut',1)
df_3 = df_3.drop('grupo',1)
df_3 = df_3.drop('Rut_Grupo',1)
```

```python
df_3['RUT'] = df_3['RUT_v']
df_3 = df_3.drop('RUT_v',1)
```

```python
df_3.head()
```

```python
# Tabla dinámica para resumir la última compra y el monto total de venta por cliente

df_4 = pd.pivot_table(df_3, index = ['RUT'], values = ['Fecha','$ Venta','$ Margen','N° Folio'],
                     aggfunc = {'Fecha': np.max, '$ Venta': np.sum, '$ Margen': np.sum}).reset_index()
```

```python
# Incluimos el día de hoy para obtener recencia

df_4['Today'] = '2021-10-01'
#df_4['Today'] = pd.Timestamp.today().strftime('%d-%m-%Y')
df_4['Today'] = pd.to_datetime(df_4.Today)
```

```python
df_4.head()
```

```python
df_4.dtypes
```

```python
df_4.count()
```

### Recencia

```python
df_4['Recencia'] = df_4['Today'] - df_4['Fecha']
```

```python
df_4.head()
```

```python
df_4['Recencia'] = df_4['Recencia'].astype('string')
df_4['Recencia'] = pd.to_numeric(df_4['Recencia'].str.split(' ').str[0])
```

```python
# Recencia Mensual

df_4['Recencia_Mensual'] = np.round(pd.to_numeric(df_4['Recencia'])/30,1)
```

```python
df_4.head()
```

```python
df_4['Recencia'].mean()
```

```python
df_4['Recencia_Mensual'].mean()
```

### Frecuencia (Periodos)

```python
# Nos quedamos sólo con las variables relevantes

df_5 = df_3[['RUT','Periodo']]
df_5.head()
```

```python
df_5.count()
```

```python
# Datos únicos

df_5 = df_5.drop_duplicates()
```

```python
df_5
```

```python
# Tabla dinámica contando el número de folios por cliente

df_5 = pd.pivot_table(df_5, index=['RUT'], values=['Periodo'], aggfunc=np.count_nonzero).reset_index()
df_5.head()
```

```python
# Renombramos la columna
df_5['Frecuencia'] = df_5['Periodo']
df_5 = df_5.drop('Periodo',1)
df_5.head()
```

```python
df_5.count()
```

### JOIN

```python
df_final = df_5.join(df_4.set_index('RUT'), on = 'RUT')
```

```python
df_final = df_final.drop('Fecha',1)
df_final = df_final.drop('Today',1)
```

```python
df_final.head()
```

```python
# Limpieza datos negativos
df_final = df_final[df_final['$ Venta'] > 10]
df_final = df_final[df_final['$ Margen'] > 10]

# Promociones por $8; Ampolletas $40
```
