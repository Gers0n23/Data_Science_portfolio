### Levantamiento 

```python
df_levantamiento = pd.read_csv(URL+'Consolidado_Top_y_Prospectos.csv', sep = ";")
```

```python
df_levantamiento.head()
```

```python
df_levantamiento.columns
```

```python
df_levantamiento = df_levantamiento[[' ppu ','RUT_Final',' nomtot ',' ¿Posee contrato de mantención? ',' ¿Está bajo leasing financiero? ']]
```

```python
df_levantamiento.columns = ['ppu','RUT_Final_2','nomtot_Lto','mantencion','leasing']
```

```python
df_levantamiento.head()
```

```python
df_parque2020 = pd.merge(df_parque2020, df_levantamiento, on = 'ppu', how = 'left')
```

```python
df_parque2020['mantencion'] = df_parque2020['mantencion'].fillna('SIN INFO')
df_parque2020['leasing'] = df_parque2020['leasing'].fillna('SIN INFO')
```

```python
# Reemplazando los rut y nombre de empresas

df_parque2020.loc[df_parque2020['RUT_Final_2'].notnull(), 'RUT_Final'] = df_parque2020['RUT_Final_2']
df_parque2020.loc[df_parque2020['nomtot_Lto'].notnull(), 'nomtot'] = df_parque2020['nomtot_Lto']

df_parque2020['RUT_Final'] = df_parque2020['RUT_Final'].astype('int64')
df_parque2020['nomtot'] = df_parque2020['nomtot'].str.upper()
```

```python
df_parque2020 = df_parque2020.drop('RUT_Final_2', axis = 1)
df_parque2020 = df_parque2020.drop('nomtot_Lto', axis = 1)
```
