```python
# RUT relacionados

df_rutgrupos = pd.read_csv(URL + 'Rut_relacionados.csv', sep = ';')
```

```python
df_rutgrupos.columns = ['rut', 'grupo','Rut_Grupo']
df_rutgrupos.head(10)
```

```python
# LEFT

df_parque2020 = pd.merge(df_parque2020, df_rutgrupos, on = 'rut', how = 'left')
```

```python
df_parque2020.head(5)
```

```python
df_parque2020.loc[df_parque2020['Rut_Grupo'].notnull(), 'RUT_Final'] = df_parque2020['Rut_Grupo']
df_parque2020.loc[df_parque2020['RUT_Final'].isnull(),'RUT_Final'] = df_parque2020['rut']
```

```python
# Eliminación columnas

df_parque2020 = df_parque2020.drop('Rut_Grupo',1)
df_parque2020 = df_parque2020.drop('grupo',1)
```
