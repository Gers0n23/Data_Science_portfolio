## Cálculo RFM

```python
quantiles = df_rfm.quantile(q = [0.25,0.50,0.75])
quantiles = quantiles.to_dict()
```

```python
def RScore(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]:
        return 2
    else:
        return 1
    
def FMScore(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]:
        return 3
    else:
        return 4
```

```python
df_rfm['R_quartile'] = df_rfm['Recencia'].apply(RScore, args = ('Recencia', quantiles))
df_rfm['F_quartile'] = df_rfm['Frecuencia'].apply(FMScore, args = ('Frecuencia', quantiles))
df_rfm['M_quartile'] = df_rfm['Valor Monetario'].apply(FMScore, args = ('Valor Monetario', quantiles))
```

```python
df_rfm.head()
```

```python
df_rfm[df_rfm['RUT_Final'] == 76588089]
```

```python
df_rfm['RFM_Score'] = df_rfm.R_quartile.map(str) + df_rfm.F_quartile.map(str) + df_rfm.M_quartile.map(str)
df_rfm['RFM_Sum'] = df_rfm[['R_quartile','F_quartile','M_quartile']].sum(axis=1)
```
