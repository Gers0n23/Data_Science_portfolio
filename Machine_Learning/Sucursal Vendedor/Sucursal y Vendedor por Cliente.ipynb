```python
import pandas as pd
```

```python
venta_path = 'C:/Users/juan.vergara/OneDrive - Caren Repuestos Flotacentro/Documentos/Análisis/Parque Automotriz/Parque en Bruto para Modelo/'

df_sucven = pd.read_csv(venta_path + 'Ventas20162021porRutSucursalVendedor.csv', sep = ';')

df_sucven.columns = ['CeCo', 'Vendedor','RUT','$ Venta']
```

```python
df_sucven.head()
```

```python
df_sucven = df_sucven[df_sucven['RUT'].notnull()]
```

```python
df_sucven['RUT'] = df_sucven['RUT'].str.split('-').str[0]
df_sucven['RUT'] = df_sucven['RUT'].astype('int64')
df_sucven['$ Venta'] = df_sucven['$ Venta'].str.replace('.','')

df_sucven['$ Venta'] = df_sucven['$ Venta'].astype('int64')

df_sucven.columns = ['CeCo','Vendedor','RUT','$ Venta']
```

```python
df_sucven.head()
```

```python
RFM_path = 'C:/Users/juan.vergara/OneDrive - Caren Repuestos Flotacentro/Documentos/Análisis/Segmentación Clientes/RFM-2022/'

df_rut = pd.read_csv(RFM_path + 'RUT_Clientes.csv')
```

```python
df_rut.head()
```

```python
df_sucven = pd.merge(df_sucven, df_rut, on ='RUT', how = 'left')

df_sucven.head()
```

```python
df_sucven[df_sucven['RUT'] == 76327820]
```

```python
# Venta para dos mejores

df_vtasxsuc2m = df_sucven[['$ Venta','RUT','CeCo']]
df_vtasxsuc2m = df_vtasxsuc2m.sort_values(by = ['$ Venta','RUT'], ascending=False)
df_vtasxsuc2m = df_vtasxsuc2m.groupby(['RUT'], as_index=False).head(2)


df_vtasxsuc2m[df_vtasxsuc2m['RUT'] == 76327820]
```

### Venta Total

```python
# Venta Total

df_vtasxsuc = df_sucven[['$ Venta','RUT']]
df_vtasxsuc = df_vtasxsuc.groupby(['RUT'], as_index=False).sum()

df_vtasxsuc.columns = ['RUT','Venta Total']
df_vtasxsuc.head()
```

```python
df_vtasxsuc[df_vtasxsuc['RUT'] == 76327820]
```

```python
# LEFT JOIN

df_final = pd.merge(df_vtasxsuc2m, df_vtasxsuc, on = 'RUT', how = 'left')
df_final['% Venta'] = df_final['$ Venta'] / df_final['Venta Total']

df_final.head()
```

```python
df_final[df_final['RUT'] == 76327820]
```

```python
df_final = df_final[df_final['% Venta'] >= 0.20]
```

```python
df_final.to_csv(venta_path+ 'VentaSucursal.csv', index = False)
```
