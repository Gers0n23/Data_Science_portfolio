Pregunta 2: Utilice el webscraping para extraer los datos de ingresos de Tesla

Muestre las últimas cinco filas del marco de  datos tesla_revenue utilizando la  función tail. Sube una captura de pantalla de los resultados.

```python
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"}

html_data = requests.get(url, headers=headers).text
soup = BeautifulSoup(html_data, 'html.parser')

tables = pd.read_html(html_data)
tesla_revenue = tables[1]

tesla_revenue.columns = ['Date', 'Revenue']
tesla_revenue['Revenue'] = tesla_revenue['Revenue'].str.replace('$', '').str.replace(',', '').str.replace('B', '').astype(float)
tesla_revenue['Revenue'] = tesla_revenue['Revenue'] * 1000 

tesla_revenue.tail()
```

Make Graph para preguntas 5 y 6

```python
def make_graph(stock_data, revenue_data, stock_name):
    fig = go.Figure()
    
    fig.add_trace(go.Scatter(x=stock_data.Date, y=stock_data.Close,
                 name="Stock Price", line_color='#0066ff'))
    
    fig.add_trace(go.Scatter(x=revenue_data.Date, y=revenue_data.Revenue,
                 name="Revenue", line_color='#ff0000'))
    
    fig.update_layout(
        title=f"{stock_name} Stock Price and Revenue",
        xaxis_title="Date",
        yaxis_title="USD",
        legend_title="Legend"
    )
    
    fig.show()
```
